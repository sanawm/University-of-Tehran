**Question 1:**

**Defining State in This Problem:**
In this warehouse game, a state represents a unique configuration of the game world at a given time. The most important dynamic elements in the game are:

Player’s Position (H) → The current coordinates of the player.
Box Positions (B1, B2, …, Bn) → The locations of all movable boxes.
Thus, the optimal state representation can be defined as:

state = (player position , tuple of box positions)
This minimal representation avoids redundant information and significantly improves efficiency.

**Problems with a Naïve State Representation (Using the Full Map):**
A basic way to represent a state is to store the entire warehouse grid at every step. However, this approach has several issues:

**1. Increased State Space & Memory Usage**
The warehouse contains static elements (walls, goals, empty spaces) that never change during the game.
Storing the full map increases the number of possible states, making search algorithms inefficient.
Every time a new state is generated, it requires copying the entire map, which is costly in terms of memory and computation.
Improvement: Store only the dynamic elements (player + box positions), significantly reducing memory usage.

**2. Slow State Comparisons**
If we store the entire grid, checking whether two states are the same requires comparing each cell in the grid.
This slows down algorithms like BFS, DFS, and A* that frequently check if a state has been visited before.
Improvement: Instead of a full grid, represent a state as a hashable tuple of coordinates.
Example: ((1,3), ((2,2), (3,4), (5,5))) → Fast comparisons using hash tables.

**3. Redundant Symmetric States**
The player’s movement around boxes without pushing them does not change the meaningful state.
If a search algorithm treats these as separate states, it explores many unnecessary paths.
Improvement:Ignore player position when detecting visited states, unless the player’s location affects box movement.
Use canonical state representations (e.g., sorting box positions) to merge equivalent states.

**4. Deadlock States Are Not Avoided**
A major issue with the naïve representation is that it doesn’t recognize deadlocks. Example:
A box is pushed into a corner (against two walls) → It can never be moved again, making the game unwinnable.
A box is aligned against a wall without a goal behind it → The state is useless.
Improvement: Implement deadlock detection to recognize unwinnable states early and prune them from the search.
Techniques include:
Corner Deadlock Detection: A box in a corner (unless a goal is there) → Invalid state.
Freeze Detection: If a box is stuck against a wall and cannot move, mark the state as unreachable.

**Final Optimized State Representation:**
Instead of storing the full grid, an optimized state contains:
Player’s Position (can be ignored for repeated state checks).
Tuple of Box Positions (for efficient storage & fast lookup).
Deadlock Detection (pruning impossible paths early).
Efficient State Representation:
state = (player_position, tuple(box_positions))
This representation significantly improves:
1. Memory Usage → No redundant static elements.
2. Speed → Fast state comparisons using hash tables.
3. Search Efficiency → Ignores irrelevant moves & detects deadlocks.

This optimized approach allows algorithms like BFS, A*, and IDA* to perform efficiently, leading to faster and more optimal solutions

**Question 5:**

Explanation of Implemented Algorithms
We have implemented several search algorithms to solve the box-pushing problem. Each algorithm explores the search space differently and has its own advantages and disadvantages.

**1. Breadth-First Search (BFS)**

BFS explores all possible moves level by level, ensuring that it finds the shortest possible solution.
It uses a queue (FIFO), expanding all possible states at a given depth before moving deeper.
Advantage: Guarantees the optimal solution (shortest path).
Disadvantage: Requires a large amount of memory, as it stores all visited states.

**2. Depth-First Search (DFS)**

DFS explores a single path as deep as possible before backtracking.
It uses a stack (LIFO) or recursion to explore states.
Advantage: Requires significantly less memory than BFS.
Disadvantage: It does not guarantee optimality and can get stuck in infinite loops in large search spaces.

**3. Iterative Deepening Search (IDS)**

IDS is a hybrid between BFS and DFS.
It runs DFS up to a depth limit, increasing the limit iteratively.
Advantage: Finds the optimal solution like BFS but uses less memory like DFS.
Disadvantage: Repeats work in early iterations, making it slower than BFS in some cases.

**4. Astar Search:**

A* is an informed search algorithm that uses a heuristic function to prioritize states.
Advantage: Fast and optimal, avoids unnecessary state expansions.
Disadvantage: Depends on the heuristic function; if the heuristic is poor, performance suffers.

**5. Weighted Astar Search:**

A variant of A* that modifies the cost function to prioritize speed:
$$
f(n)=g(n)+w⋅h(n
$$
Advantage: Faster than A* when the weight is set appropriately.
Disadvantage: May not always find the optimal solution (depends on weight value).

**Differences and Advantages of Each Algorithm:**
| Algorithm      | Optimal Solution | Memory Usage     | Speed       | Special Features |
|--------------|----------------|-----------------|------------|------------------|
| **BFS**     | ✅ Yes         | ❌ High        | ⏳ Slow    | Guarantees shortest path |
| **DFS**     | ❌ No          | ✅ Low         | ⚡ Fast    | Gets stuck in deep loops |
| **IDS**     | ✅ Yes         | ✅ Low         | ⏳ Medium  | Combines BFS optimality with DFS memory efficiency |
| **A***      | ✅ Yes         | ⚠️ Medium-High | ⚡ Fast    | Uses heuristics to speed up search |
| **Weighted A*** | ⚠️ Not Always | ⚠️ Medium-High | ⚡⚡ Very Fast | Can sacrifice optimality for speed |



**(a) What is the problem with the DFS algorithm? Why is this algorithm still used despite this?**

Problems with DFS:
No guarantee of optimality → DFS does not necessarily find the shortest solution because it explores deep paths first.
Can get stuck in loops → If cycles exist in the search space, DFS might keep revisiting the same states indefinitely.
Deep search wastes time → DFS may go down long incorrect paths before backtracking.

Why is DFS still used?
✅ Low memory usage → DFS only needs to store the current path, unlike BFS, which stores all visited states.
✅ Fast for deep solutions → If the solution is deep, DFS can reach it quickly without expanding unnecessary nodes.
✅ Works well with iterative deepening (IDS) → When combined with IDS, it gains BFS’s optimality while maintaining DFS’s memory efficiency.

**(b) How does IDS combine BFS and DFS, and what problems does it solve?**

How IDS Works:
IDS runs DFS with a depth limit and gradually increases the depth until the solution is found.
It essentially mimics BFS level expansion while keeping DFS’s memory efficiency.
Problems IDS Solves
✅ Solves DFS’s infinite loop issue → By limiting depth, IDS prevents DFS from getting stuck.
✅ Guarantees an optimal solution → Like BFS, IDS explores all states at each depth before going deeper.
✅ Uses much less memory than BFS → Instead of storing all visited states, IDS only keeps track of the current search path.

Trade-off of IDS: 
Redundant computation → IDS repeatedly explores early levels, making it slightly slower than BFS in some cases.

**Final Summary:**
BFS → Guarantees optimal solution but uses a lot of memory.
DFS → Uses less memory but does not guarantee optimality.
IDS → Balances BFS and DFS, finding optimal solutions with less memory.
A* → Uses heuristics to speed up search while keeping optimality.
Weighted A* → Faster than A* but may sacrifice optimality.
1. For optimal solutions: Use BFS, IDS, or A*.
2. For speed: Use Weighted A* with a good heuristic.
3. For memory efficiency: Use IDS or DFS.
